```python
my_list = list([1, 2, 'abc', 12.13])
my_list.append('Python')
print(my_list)
my_list.append([11, 12, 13])
print(my_list)
```

```python
my_list = [1, 2, 'abc', 1.2]
my_list.insert(3, 5)
print(my_list)
my_list.insert(4, [11, 12, 13])
print(my_list)
```

```python
my_list = list([1, 2, 'abc', 4.5])
my_list.extend([11, 12, 13])
print(my_list)
```

```python
my_list = list([2, 4, 6, 8, 6, 10, 12])
my_list.remove(6)
my_list.remove(8)

print(my_list)
```

```python
my_list = list([2, 4, 6, 8, 10, 12])

my_list.pop(2)
print(my_list)

my_list.pop()
print(my_list)
```

```python
my_list = list([2, 4, 6, 8, 10, 12])
del my_list[2:5]
print(my_list)

my_list = list([1, 2, 3, 4, 5, 6])
del my_list[3:]
print(my_list)
```
