# Combined conditional statement and iteration


```python
x = [1,2,3,4,5]
```

```python
for number in x:
    if number > 3:
        print("greater than 3")
    else:
        print("less than or equal 3")
```

```python
y = [0,2,4,6,8,11]
```

```python
for number in y:
    if number % 2 == 0:
        print("even")
    else:
        print("odd")
```
