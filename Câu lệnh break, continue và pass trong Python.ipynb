```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
dem=0
for i  in range (a,b+1):
    if i%2!=0:
        if i%5==0 or i%7==0:
            continue
        dem=dem+1
print("Các số lẻ: ",dem)
```
