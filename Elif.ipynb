# Conditional statement: elif


```python
def greater_than_million(y):
    if y >= 1000000:
        return ("High class")
    elif 500000 < y < 1000000:
        return ("Middle class")
    else:
        return ("Normal")
```

```python
greater_than_million(2000000)
```

```python
greater_than_million(200000)
```

```python
greater_than_million(700000)
```
