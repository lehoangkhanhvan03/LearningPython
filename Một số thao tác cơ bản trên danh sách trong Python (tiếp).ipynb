```python
lopA1=[7.8,5.6,8.7,8.9,9,9.5]
lopA2=[6.0,6.5,9.3,9.2,7.5]

choice=input("Hãy lựa chọn yêu cầu:\n1. Bấm phím 1 nếu muốn sắp xếp bảng điểm theo thứ tự tăng dần.\n
2. Bấm phím 2 nếu muốn sắp xếp bảng điểm theo thứ tự giảm dẫn.\n")

if choice == "1":
  truong = lopA1 + lopA2
  truong.sort()
  print(truong)

if choice == "2":
  truong = lopA1 + lopA2
  truong.sort(reverse = True)
  print(truong)
```

```python
fruits = ["apple", "banana", "cherry", "date"]

# Sắp xếp theo độ dài của các chuỗi
fruits.sort(key=len)
print(fruits)  # Kết quả: ['date', 'apple', 'banana', 'cherry']

# Sắp xếp theo thứ tự ngược (giảm dần)
fruits.sort(reverse=True)
print(fruits)  # Kết quả: ['date', 'cherry', 'banana', 'apple']
```
