### Cách 1


```python
import yfinance as yf
import pandas as pd

# Tải dữ liệu của một mã cổ phiếu, VD: Apple (AAPL)
stock_code = 'AAPL' 
stock_data = yf.download(stock_code, start='2022-01-01', end='2023-01-01')

# Hiển thị dữ liệu
print(stock_data.head())
```

### Cách2

```python
import yfinance as yf
# Request historical data for past 5 years
data = yf.Ticker("NVDA").history(period='5y')
# Show info
print(data.info())
```

### Cách 3


```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ giá đóng cửa hàng ngày
stock_data['Adj Close'].plot() #.plot() được gọi để vẽ biểu đồ dữ liệu này
plt.title('Biểu đồ giá đóng cửa hàng ngày của {}'.format(stock_code)) 
#plt.title(). Biểu đồ sẽ có tiêu đề là 'Biểu đồ giá đóng cửa hàng ngày của AAPL' 
#stock_code = 'AAPL' đã khai báo ở trên
plt.xlabel('Ngày') #trục x(hoành độ) là Ngày
plt.ylabel('Giá') #trục y(tung độ) là Giá
plt.show() #hiển thị biểu đồ đã được tạo ra từ dữ liệu
```
