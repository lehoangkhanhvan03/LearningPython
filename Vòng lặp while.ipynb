```python
s = 0
while (True):
    num = int(input("Nhập số:"))
    if num < 0:
        break
    else:
        s = s + num
print("Tổng bằng: ",s)   
```

```python
arr = []
while (True):
    num = int(input("Nhập số:"))
    if num < 0:
        break
    elif num % 2 != 0:
        continue
    else:
        arr.append(num)
print("In ra màn hình:",arr)
        
```
