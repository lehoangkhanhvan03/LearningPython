```python
string=input()

str_largest = ""

for s in string.split():
  if len(s) > len(str_largest):
    str_largest = s

print("The longest string is \"" + str_largest + "\" with", len(str_largest), "characters")
```

```python
def convert_text(n):
  word = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
  text = []
  for num in str(n):
    text.append(word[int(num)])
  return " ".join(text)
  
n = int(input())

print(convert_text(n))
```
